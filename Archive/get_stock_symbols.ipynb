```python
import os
import glob
import requests
import configparser
import json
import pandas as pd
```

```python
config = configparser.ConfigParser()
CONFIG_FILE = 'dwh.cfg'
config.read(CONFIG_FILE)

TOKEN   = config.get("API", 'TOKEN')
DOMAIN  = config.get('API', 'DOMAIN')
VERSION = config.get('API', 'VERSION')
```

```python
METAFILE = 'symbols_valid_meta.csv'
all_symbols = []
for root, dirs, files in os.walk('stock-market-dataset'):
    for file in files:
        if file != METAFILE:
            symbol = file.split('.')[0]
            all_symbols.append(symbol)
```

```python
API_HTTPS = f"https://{DOMAIN}/{VERSION}/stock/AA/company?token={TOKEN}"
response = requests.get(API_HTTPS)
```

```python
d = []
errors = []
SUCCESS_CODE = 200
for symbol in all_symbols:
    print(f"INFO: Fetching {symbol}")
    api_https = f"https://{DOMAIN}/{VERSION}/stock/{symbol}/company?token={TOKEN}"
    response = requests.get(api_https)
    if response.status_code == SUCCESS_CODE:
        d.append(json.loads(response.text))
        print("Done!")
    else:
        print(f"Error on {symbol}. Skipping")
        errors.append((response.status_code, symbol))
```

```python
d = []
errors = []
SUCCESS_CODE = 200
for symbol in all_symbols:
    print(f"INFO: Fetching {symbol}")
    api_https = f"https://{DOMAIN}/{VERSION}/stock/{symbol}/advanced-stats?token={TOKEN}"
    response = requests.get(api_https)
    if response.status_code == SUCCESS_CODE:
        json_data = json.loads(response.text)
        json_data['symbol'] = symbol
        d.append(json_data)
        
        print("Done!")
    else:
        print(f"Error on {symbol}. Skipping")
        errors.append((response.status_code, symbol))
```
